Let's next check out Yunhui's transition count routine:

````python
def construct_raw_tcm(assignment, tau, nstates):
    """Constructs a count matrix from the assignments."""
    
    # determine if it is a numpy array or a file name to load

    if type(assignment) == str:
        ass = np.load(assignment)
    else:
        ass = assignment
        
    # check if this is one trajectory or multiple trajectories
    if len(ass) == 1:
        tcm = np.zeros((nstates, nstates))
        for j in range(len(ass)-tau):
            tcm[ass[j],ass[j+tau]]+=1
    else:
        tcm= np.zeros((nstates, nstates)) # make transition count matrix nstates x nstates
        for i in range(len(ass)): #loop through all traj
            print('Collecting counts from traj', i)
            if len(ass[i]) > tau:  # check if traj length is longer than the lagtime
                for j in range(len(ass[i])-tau):  # sliding windows
                    tcm[ass[i][j],ass[i][j+tau]] += 1  # add transition count
            else:
                pass
    return tcm


tau = 10
C = construct_raw_tcm(a, tau, 18)
print('C', C.astype(int))
````

````python
print('C', C.astype(int))

nonsym_residuals = C - C.transpose()
print(nonsym_residuals.astype(int))
````

````python
from estimators3 import *
T_rownorm = rownormalized_tProb(C, symmetrize=True, verbose=True)

for i in range(nstates):
    print('T_rownorm[i,:].sum()', T_rownorm[i,:].sum())

````

````python
from numpy.linalg import eig

print("computing eigenvalues and eigenvectors...")
mu, eigenvectors = eig(T_rownorm.transpose())    # compute eigenvalues and eigenvectors

# Sort the eigenvectors by their computed eigenvalues
Ind = np.argsort(-mu)
mu, eigenvectors = mu[Ind], eigenvectors[:,Ind]

# fix the sign structure of the evecs, if needed
if eigenvectors[:,0].sum() < 0.0:
    eigenvectors = eigenvectors * -1.0


plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(range(nstates), mu, 'o-')
plt.xlabel('rank')
plt.ylabel('eigenvalue')
plt.yscale('log')
plt.tight_layout()


tau_in_ns = 0.001*tau   
implied_timescales = [-1.0*tau_in_ns/np.log(mu[i]) for i in range(1,nstates)]    

plt.subplot(1,2,2)
plt.plot(range(1,nstates), implied_timescales, 'o-')
plt.xlabel('rank')
plt.ylabel('implied timescale (ns)')
plt.ylim(0.001,100)
plt.yscale('log')

plt.tight_layout()
````
